# Data visualization

```python
df.columns
```

```python
#Revenue Contribution by Product Category
#scatterplot
sns.scatterplot(data=df, x="Product Category", y="Total Revenue",color='g')
plt.title("Revenue Contribution by Product Category")
plt.xticks(rotation=45)
plt.show()
```

```python
#Revenue Contribution by Product Category
#lineplot
sns.lineplot(data=df, x="Product Category", y="Total Revenue",color='r')
plt.title("Revenue Contribution by Product Category")
plt.xticks(rotation=45)
plt.show()
```

```python
#Revenue contribution by category
#barplot
top_products = df.groupby("Product Name").sum().sort_values("Total Revenue", ascending=False).head(20)
top_products.plot(kind='bar', y='Total Revenue', legend=False, color='b')
plt.title("Top 10 Revenue-Generating Products")
plt.ylabel("Total Revenue")
plt.show()
```

```python
#Revenue by region
#barplot
sns.barplot(data=df, x="Region", y="Total Revenue",color='k')
plt.title("Revenue by Region")
plt.xlabel("Region")
plt.ylabel("Total Revenue")
plt.show()
```

```python
#Revenue over time
#lineplot
time_data = df.groupby("Date").sum().reset_index()
sns.lineplot(data=time_data, x="Date", y="Total Revenue",color='pink')
plt.title("Revenue Trend Over Time")
plt.ylabel("Total Revenue")
plt.show()
```

```python
# Product Category Revenue Over Time
#lineplot
sns.lineplot(data=df, x="Product Category", y="Units Sold")
plt.title("Revenue by Product Category Over Time")
plt.ylabel("Total Revenue")
plt.legend(loc="upper left")
plt.show()
```

```python
# Histogram of Total Revenue
sns.histplot(data=df, x="Total Revenue", bins=30, kde=True)
plt.title("Distribution of Total Revenue")
plt.show()
```

```python
# Correlation Heatmap
correlation = df.corr(numeric_only=True)
sns.heatmap(correlation, annot=True, cmap="cool", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Rug plot
sns.rugplot(data=df, x="Total Revenue",hue="Product Category")
plt.title("Rug Plot of Total Revenue by Product Category")
plt.show()
```

```python
#Pairplot
sns.pairplot(data=df, vars=["Total Revenue", "Units Sold", "Unit Price"], hue="Product Category")
plt.show()
```

```python
#Regression plot
sns.regplot(data=df, x="Units Sold", y="Total Revenue", line_kws={"color": "red"})
plt.title("Regression Plot: Total Revenue vs Units Sold")
plt.show()
```

```python
 #Distribution of Unit Price
sns.histplot(data=df, x="Unit Price", kde=True, bins=20)
plt.title("Distribution of Unit Price")
plt.xlabel("Unit Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# Strip Plot: Total Revenue by Product Category
sns.stripplot(data=df, x="Product Category", y="Total Revenue", jitter=True, palette="viridis")
plt.title("Strip Plot: Total Revenue by Product Category")
plt.xticks(rotation=45)
plt.show()
```

```python
# Joint Plot: Total Revenue vs Units Sold
sns.jointplot(data=df, x="Units Sold", y="Total Revenue", kind="scatter", hue="Product Category")
plt.suptitle("Joint Plot: Total Revenue vs Units Sold", y=1.02)
plt.show()
```

```python
# Violin Plot: Total Revenue by Product Category
sns.violinplot(data=df, x="Product Category", y="Total Revenue", palette="muted")
plt.title("Violin Plot: Total Revenue by Product Category")
plt.xticks(rotation=45)
plt.show()
```

```python
# FacetGrid: Total Revenue by Product Category and Region
g = sns.FacetGrid(df, col="Region", hue="Product Category", height=4, col_wrap=3)
g.map(sns.scatterplot, "Units Sold", "Total Revenue", alpha=0.7)
g.add_legend()
plt.suptitle("Facet Grid: Total Revenue by Region and Product Category", y=1.02)
plt.show()
```

```python
# Strip Plot: Total Revenue by Product Category
sns.stripplot(data=df, x="Product Category", y="Total Revenue",palette="coolwarm")
plt.title("Swarm Plot: Total Revenue by Product Category")
plt.xticks(rotation=45)
plt.show()
```

```python
#Matplotlib
#scatterplot
plt.figure(figsize=(10, 6))
plt.scatter(df["Units Sold"], df["Total Revenue"], marker='*', color='g', alpha=0.7, label='Revenue vs Units Sold')
plt.title("Scatter Plot: Total Revenue vs Units Sold")
plt.xlabel("Units Sold")
plt.ylabel("Total Revenue")
plt.show()

```

```python
# Group data by Region and sum up Units Sold
region_df = df.groupby("Region").sum()["Units Sold"]

plt.figure(figsize=(10, 6))
colors = ['g','r','b']
plt.bar(region_df.index, region_df, color=colors, edgecolor='black', linewidth=1.5)
plt.title("Units Sold by Region")
plt.xlabel("Region")
plt.ylabel("Units Sold")
plt.show()
```

```python
# Group data by Product Category and sum up Total Revenue
category_data = df.groupby("Product Category").sum()["Total Revenue"]

plt.figure(figsize=(8, 8))
colors = ['gold', 'lightblue', 'lightgreen', 'lightcoral', 'pink']
plt.pie(category_data, labels=category_data.index, colors=colors, autopct='%1.1f%%', startangle=100, shadow=True)
plt.title("Revenue Distribution by Product Category")
plt.show()
```

# Conclusion
The project aimed to explore and analyze the Total Revenue trends and identify key factors that contribute to revenue generation. Using various visualizations, including line plots, bar charts, scatter plots, and pie charts, we were able to uncover valuable patterns, relationships, and anomalies in the data.
The analysis suggested that factors such as product category, region, and the number of units sold directly influence the total revenue. A deeper look into the pricing strategy and discounting might also help optimize revenue generation.
A small number of categories, such as "Electronics" and "Home Appliances," dominate total revenue.Categories like "Groceries" and "Clothing" generate significant sales volumes but may contribute less to total revenue due to lower average prices. Within the leading categories, specific products (e.g., Smartphone or Refrigerator ) significantly impact revenue.
